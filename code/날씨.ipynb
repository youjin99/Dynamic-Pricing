```python
import numpy as np
import pandas as pd
import pickle
```

```python
with open("/content/drive/Shareddrives/인빅_프로젝트/FC주문/FC_order.pickle","rb") as fr:
    FC_order = pickle.load(fr).reset_index(drop=True)

FC_order
```

```python
dat = FC_order[['INS_DATE',]]
dat['CNEE_ADDR'] = pd.DataFrame(FC_order['CNEE_ADDR_1'] + ' ' + FC_order['CNEE_ADDR_2'])
dat.head(1)
```

```python
w = pd.read_csv('/content/drive/Shareddrives/인빅_프로젝트/전민기/OBS_ASOS_DD_20211023204742.csv', encoding='euc-kr')
w.head()
```

```python
for i in range(len(w)):
  w.loc[i, '일시'] = w.loc[i, '일시'].replace('-', '') 

w = w.drop(['지점'], axis=1)
```

```python
w = w.fillna(0)
```

```python
INS_DATE = list(dat['INS_DATE'])
CNEE_ADDR = list(dat['CNEE_ADDR'].fillna('서울'))
일시 = list(w['일시'])
지점명 = list(w['지점명'])
강수량 = list(w['일강수량(mm)'])
```

```python
for i in range(len(CNEE_ADDR)):
  if '16817' in CNEE_ADDR[i]:
    CNEE_ADDR[i] = '수원'
```

```python
f = list()
for i in set(CNEE_ADDR):
  for j in set(지점명):
    if j in i: # 지점명 in CNEE_ADDR
      f.append(i)

set(CNEE_ADDR) - set(f) # CNEE_ADDR - 지점명
```

```python
속초 = ['고성', '양양']
태백 = ['삼척']
인제 = ['양구']
대관령 = ['평창']
철원 = ['화천']
원주 = ['횡성', '태종로']
파주 = ['고양', '의정부',]
수원 = ['과천', '군포', '안산', '안성', '성남', '오산', '용인', '의왕', '평택', '화성']
서울 = ['구리', '남양주', '안양', '하남', '광명', '김포', '마포']
이천 = ['여주', '경기도 여주시']
인천 = ['부천', '시흥']
동두천 = ['양주', '연천', '포천', '가평']
통영 = ['고성', '사천', '하동']
대구 = ['경산', '고령', '청도', '칠곡']
구미 = ['김천', '성주', '군위', '예천', '영양']
순천 = ['곡성', '구례']
광주 = ['나주', '담양', '화순', '함평', '장성']
목포 = ['무안', '신안', '영암']
군산 = ['김제']
전주 = ['완주', '익산', '진안']
부여 = ['계룡', '공주', '논산', '무주']
서산 = ['당진', '아산', '예산']
보령 = ['서천', '태안', '청양']
제천 = ['단양']
대전 = ['영동', '옥천']
충주 = ['음성', '증평', '진천', '괴산']
의령 = ['함안', '의령', '창녕']
울릉 = ['울릉', '경상북도 울릉군']
```

```python
a = [속초, 태백, 인제, 대관령, 철원, 원주, 파주, 수원, 서울, 인천, 동두천, 통영, 대구, 의령, 구미, 순천, 광주, 목포, 군산, 전주, 부여, 서산, 보령, 제천, 대전, 충주, 의령, 울릉, 이천]
b = ['속초', '태백', '인제', '대관령', '철원', '원주', '파주', '수원', '서울', '인천', '동두천', '통영', '대구', '의령', '구미', '순천', '광주', '목포', '군산', '전주', '부여', '서산', '보령', '제천', '대전', '충주', '의령', '울릉', '이천']

for i in range(len(a)):
  for j in a[i]:
    for k in range(len(w)):
      if j in 지점명[k]:
        지점명[k] = b[i]
        break
```

```python
for i in range(len(a)):
  for k in range(len(CNEE_ADDR)):
    for j in a[i]:
      if j in CNEE_ADDR[k]:
        CNEE_ADDR[k] = b[i]
        break
```

```python
g = list()
for i in set(CNEE_ADDR):
  for j in set(지점명):
    if j in i: # 지점명 in CNEE_ADDR
      g.append(i)

set(CNEE_ADDR) - set(g) # CNEE_ADDR - 지점명
```

```python
c = list()

for i in range(len(INS_DATE)):
  for j in range(len(w)):
    if int(INS_DATE[i]) + 1 == int(일시[j]):
      if 지점명[j] in CNEE_ADDR[i]:
        c.append(float(강수량[j]))
        break
  else:
    c.append(0)
  print(i)
```

```python
FC_order['일강수량(mm)'] = c
FC_order.head(3)
```

```python
aa = list()
lst = list(FC_order['일강수량(mm)'])

for i in range(len(lst)):
  if lst[i] > 0:
    aa.append(int(1))
  else:
    aa.append(int(0))

FC_order['rain'] = aa # 비오면 1 안오면 0
FC_order.head(1)
```
